### 取特徵
- model name: all-MiniLM-L6-v2

```python
true_data['label'] = True
fake_data['label'] = False

data = pd.concat([true_data, fake_data], ignore_index=True)
print(data.head())
```

```python
# need to load a model at the same time
import functions.embedding as fe
```

```python
for row in data.iterrows():
    title = row[1]['title']
    text = row[1]['text']
    time = row[1]['date']
    combine_text = f"{title}[SEP] {time}[SEP] {text}"
    print(f"{combine_text}\r\n")
    (keywords, key_sentences)=fe.extract_keywords(combine_text)
    print(f"Keywords: {keywords}")
    print(f"Key Sentences: {key_sentences}")
    break
```
